```python
#Fecha: 24 de mayo de 2026
#Autor: Elizabeth Vega Pacheco
import pandas as pd
import random
```

```python
# Datos base para el dataset de Instagram 
datosInstagram = {
    "influencer": ["nomada_digital", "foodie_viajero", "mexico_en_lente", "carma_travels", "chef_playero"],
    "likes": [15400, 8900, 32100, 4200, 11300],
    "comentarios": [720, 410, 1580, 190, 540],
    "fecha": ["2025-11-04", "2026-01-15", "2026-02-28", "2026-04-10", "2026-05-25"],
    "tipo_post": ["carrusel", "reel", "foto", "story", "reel"]
}
```

```python
dfDatos = pd.DataFrame(datosInstagram)

cantidad = 1000000

# Crear el dataFrame de 1000000 de registros
dfMillon = pd.DataFrame({
    "influencer": [random.choice(dfDatos["influencer"]) for _ in range(cantidad)],
    "likes": [random.randint(100, 500000) for _ in range(cantidad)],
    "comentarios": [random.randint(10, 50000) for _ in range(cantidad)],
    "fecha": [random.choice(dfDatos["fecha"]) for _ in range(cantidad)],
    "tipo_post": [random.choice(dfDatos["tipo_post"]) for _ in range(cantidad)]
})
```

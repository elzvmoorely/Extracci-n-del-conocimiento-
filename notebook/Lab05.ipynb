```python
from faker import Faker
```

```python
fake = Faker("es_MX")
```

```python

datos = []
for i in range (100):
    datos.append(
        { 
        "id_usuario": i+1,
        "nombre": fake.name(),
        "correo": fake.email(),
        "edad": random.randint(18,65),
        "seguidores": random.randint(50,5000),
        "plataforma": random.choice(["tiktok","instagram","youtube"])
        }
    )
```

```python
dfInfluencer=pd.DataFrame(datos)
```

```python
#graficar
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize = (8,4))
sns.histplot(dfInfluencer["edad"], bins=10)
plt.title("distribucion de edades")
plt.show()
```

```python
plt.figure(figsize = (8,4))
sns.countplot(x="plataforma", data= dfInfluencer)
plt.title("cantidad de usuarios por plataforma")
plt.show()
```

## Interpretación de Resultados

Los datos generados representan perfiles ficticios con características similares a las encontradas en entornos reales.

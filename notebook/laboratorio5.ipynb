```python
import pandas as pd
import numpy as np
import random
```

```python
from faker import Faker
```

```python
pip install Faker
```

```python
fake = Faker("es_MX")
```

```python

datos = []
for i in range (100):
    datos.append(
        { 
        "id_usuario": i+1,
        "nombre": fake.name(),
        "correo": fake.email(),
        "edad": random.randint(18,65),
        "seguidores": random.radint(50,5000),
        "plataforma": random.choice(["tiktok","instagram","youtube"])
        }
    )
```

```python
dfInfluencer=pd.DataFrame(datos)
```

```python
plt.figure(figsize = (8,4))
sns.histplot(dfInfluencer["edad"], bins=10)
plt.tittle("distribucion de edades")
plt.show()
```

```python
plt.figure(figsize = (8,4))
sns.countplot(x="plataforma", data= dfInfluencer)
plt.title("cantidad de usuarios por plataforma")
plt.show()
```

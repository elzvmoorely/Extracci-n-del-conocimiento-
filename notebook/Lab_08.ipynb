```python
#IMPORTAMOS LAS VARIABLES 
import pandas as pd
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import (
    accuracy_score,
    confusion_matrix,
    classification_report,
    ConfusionMatrixDisplay
)
```

```python
#CARGAMOS EL DATASET
```

```python
iris = load_iris(as_frame=True)
```

```python
print(iris.DESCR)
```

```python
# Convertimos los datos en DataFrame
dfIris = iris.frame
dfIris.head(2)
```

```python
dfIris.tail(2)
```

```python
iris.target_names
```

```python
#Cambiamos la columna de Target por Especie
dfIris.rename(columns={"target" : "species"}, inplace=True)
```

```python
##variables predictorias 
```

## Variables predictoras y Variable objetivo

X (Variables predictoras) son aquellas que me permiten hacer la prediccion
y (objetivo) Lo que quiero predecir

```python
#Variables predictoras
X = dfIris.drop(columns=["species"])
```

```python
# Variable objetivo
y = dfIris["species"]
```

```python
print("---Variables predictoras---")
print(X.columns)
print("---Variable Objetivo ----")
print(y.name)
```

## División del conjunto de datos

*Training Set:* Datos para entrenamiento.

*Testing Set:* Datos no vistos por el modelo, para prueba.

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, #Variables predictoras
    y, #Variable objetivo,
    test_size = 0.20,
    random_state=42,
    stratify=y
)
```

```python
print("X_train: ", X_train.shape)
print("X_test: ", X_test.shape)
print("--------------------------")
print("y_train: ", y_train.shape)
print("y_test: ", y_test.shape)
```

# Escalado de las variables

```python
scaler = StandardScaler()
```

```python
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
print("-----------Datos sin escalar------------")
print(X_train.head(5))
print("-----------Escalados------------")
print(X_train_scaled[:5])
```

# Entrenamiento del modelo

```python
# Creamos una instancia del algoritmo
modelo = KNeighborsClassifier(
    n_neighbors = 3 #numero de vecinos a comparar
)
```

```python
# Entrenamos con los datos de entrenamiento
modelo.fit(
    X_train_scaled,
    y_train
)
```

# Prediccion con los datos de prueba

```python
prediccion = modelo.predict(X_test_scaled)
```

```python
pd.DataFrame({
    "Especie real" : y_test.values,
    "Especie predicha" : prediccion
}).head(10)
```
